```python
import keras_cv
from tensorflow import keras
import matplotlib.pyplot as plt

keras.mixed_precision.set_global_policy("mixed_float16")

model = keras_cv.models.StableDiffusion(jit_compile=True, img_width=512, img_height=512)

def plot_images(images):
  plt.figure(figsize=(20, 20))
  for i in range(len(images)):
    ax = plt.subplot(1, len(images), i + 1)
    plt.imshow(images[i])
    plt.axis("off")

def generate(prompt=""):
  images = model.text_to_image(prompt, batch_size=1)
  plot_images(images)
```

```python
generate("A cartoon raccoon missing a co-worker and questioning past decisions")
```
